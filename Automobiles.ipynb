```python
import pandas as pd

```

```python
datos = pd.read_csv('automobile_data.csv')
datos.head()
```

```python
colum = datos.columns
print(colum)
```

```python
nombre_col = {
    'make' : 'marca',
    'fuel_type':'tipo_combustible',
    'num_of_doors':'num_puertas',
    'body_style':'tipo',
    'drive_wheels':'traccion',
       'engine_location':'ubicacion_motor',
    'wheel_base':'distancia_entre_ejes',
    'length':'largo',
    'width':'peso',
    'height':'altura',
       'curb_weight':'peso_vacio',
    'engine_type':'tipo:motor',
    'num_of_cylinders':'num_cilindros',
    'engine_size':'tamaño_motor',
       'fuel_system':'sist_combustible',
     'compression_ratio':'ratio_compresion',
    'horsepower':'hp',
    'city_mpg':'millas_por_galonCiudad',
       'highway_mpg':'millas_por_galonRuta',
     'price':'precio'
}
```

```python
datos = datos.rename(columns=nombre_col)
datos.head()
```

```python
#convertir gas a nafta
datos.tipo_combustible.unique()
```

```python
t_comb = {'gas':'nafta', 'diesel':'diesel'}
```

```python
datos.tipo_combustible = datos.tipo_combustible.map(t_comb)
datos.sample(2)
```

```python
ibi_m = {'front':'frente', 'rear':'atras'}
```

```python
datos.ubicacion_motor = datos.ubicacion_motor.map(ibi_m)
datos.sample(3)
```

```python
datos.num_puertas.unique()
```

```python
datos.ubicacion_motor.unique()
datos.traccion.unique()
```

```python
traccion = {'rwd':'trasera', 'fwd':'delantera', '4wd ':'4x4', '4wd':'4x4'}
```

```python
datos.traccion = datos.traccion.map(traccion)
datos.sample(3)
```

```python
datos.num_puertas.unique()
```

```python
puerta= {'two':'2', 'four':'4', '':'4'}
```

```python
datos.num_puertas = datos.num_puertas.map(puerta)
datos.sample(3)
```

```python
datos['distancia_entre_eje(mt)'] = round(datos['distancia_entre_ejes'] * 0.0254,2)
datos.sample(3)
```

```python
datos['largo(mt)'] = round(datos['largo'] * 0.0254,2)
datos.sample(3)
```

```python
datos['peso(kg)'] = round(datos['peso'] * 29.9371,2)
datos.sample(3)
```

```python
datos['altura(mt)'] = round(datos['altura'] * 0.0254,2)
datos.sample(3)
```

```python
datos['km/lt(ciudad)'] = round(datos['millas_por_galonCiudad'] * 2.35215,2)
datos.sample(3)
```

```python
datos['km/lt(ruta)'] = round(datos['millas_por_galonRuta'] * 2.35215,2)
datos.sample(3)
```

```python
datos = datos.drop(['distancia_entre_ejes','largo','peso','altura','millas_por_galonCiudad','millas_por_galonRuta'],axis = 1)
datos.sample(3)
```
